```python
import pandas as pd
# read the dataset
df1=pd.read_csv("Credit_card.csv")

df2=pd.read_csv("Credit_card_label.csv")


#merging the dataset
cc=pd.merge(df1,df2, how="outer",on="Ind_ID")
import duckdb
conn=duckdb.connect()
df=cc.copy()
```

```python
conn.register("df",df)
conn.execute("select * from df").fetchdf().head(2)
```

```python
# 1. Group the customers based on their income type and find the average of their annual income.
conn.execute("select Type_Income ,avg(Annual_income) from df group by Type_Income").fetchdf().head(10)
```

```python
# 2.Find the female owners of cars and property.
result = conn.execute("SELECT * FROM df WHERE GENDER = 'F' AND Car_Owner = 'Y' AND Propert_Owner = 'Y'")
rows = result.fetchall()
for row in rows[:]:
    print(row)

```

```python
#3. Find the male customers who are staying with their families.
result = conn.execute("SELECT * FROM df WHERE GENDER = 'M' AND Housing_type = 'With parents'")
rows = result.fetchall()
for row in rows[:]:
    print(row)
```

```python
#4. Please list the top five people having the highest income.
result = conn.execute("SELECT * FROM df ORDER BY Annual_income desc limit 5")
rows = result.fetchall()
for row in rows[:]:
    print(row)

```

```python
#5. How many married people are having bad credit?
result = conn.execute("SELECT * FROM df WHERE Marital_status = 'Married' AND label = 1")
rows = result.fetchall()
for row in rows[:]:
    print(row)
```

```python
#6. What is the highest education level and what is the total count?


result = conn.execute("SELECT EDUCATION, COUNT(*) FROM df group by EDUCATION having EDUCATION= 'Academic degree'")

rows = result.fetchall()
for row in rows[:]:
    print(row)
```

```python
# 7. Between married males and females, who is having more bad credit?
result = conn.execute("""
    SELECT
        GENDER,
        COUNT(CASE WHEN GENDER = 'M' THEN 1 END) AS Male_Bad_Credit_Count,
        COUNT(CASE WHEN GENDER = 'F' THEN 1 END) AS Female_Bad_Credit_Count
    FROM
        df
    WHERE
        Marital_status = 'Married'
        AND label = 1
    GROUP BY
        GENDER
""")
rows = result.fetchall()
for row in rows[:]:
    print(row)
```
